This tutorial gives a more details about different types of recurrent neural network that can be used for MultiVariate Times-Series Data along with combinations of CNN-RNNs that can improve the performance specifically for classification tasks.

```python
#Import sktime
import sktime
#Import datasets currently available in sktime
from sktime.datasets import load_basic_motions
```

```python
# Splitting the same file into training and testing set
X_train, y_train = load_basic_motions(split="train", return_type="numpy3D")
X_test, y_test = load_basic_motions(split="test", return_type="numpy3D")
```

```python
# Sanity check to find training features and truth labels
print(X_train.shape)
print(y_train.shape)
```

**MultiVariate Time Series Transformer Classifier**

```python
#Import MultiVariate Time-Series Transformer Classifier
from sktime.classification.deep_learning import MVTSTransformerClassifier
```

```python
#Define the Transformer Classifier
mvts = MVTSTransformerClassifier(num_epochs=20, batch_size=8)
mvts.fit(X_train, y_train)
```

```python
# nested parameter interface via get_params, set_params
mvts.get_params()
```

```python
# We inspect fitted parameters if we like
mvts.get_fitted_params()
```

```python
#Store the predictions over testing set
y_pred = mvts.predict(X_test)
```

```python
# Evaluate the results
from sklearn.metrics import classification_report
```

```python
#Print evaluation metrics
print(classification_report(y_test, y_pred))
```
